```python
import numpy as np
import pandas as pd
from csv import DictReader
```

```python
vozila = DictReader(open('NIO_podatki_vozila_1.csv', 'rt', encoding='ANSI'), delimiter=""";""")

data = []

i = 0
for row in vozila:
    if (row["J-Kategorija in vrsta vozila (opis)"] != "osebni avtomobil"):
        continue
    
    newRow = {}
    
    newRow["B-Datum prve registracije vozila"] = row["B-Datum prve registracije vozila"]
    newRow["2A-Datum prve registracije vozila v SLO"] = row["2A-Datum prve registracije vozila v SLO"]
    newRow["4A-Registrsko obmocje tablice prve registracije"] = row["4A-Registrsko obmocje tablice prve registracije"]
    newRow["C-Starost uporabnika vozila"] = row["C-Starost uporabnika vozila"]
    newRow["C-Ali je uporabnik pravna ali fizicna oseba"] = row["C-Ali je uporabnik pravna ali fizicna oseba"]
    newRow["C-Ali je uporabnik tudi lastnik vozila"] = row["C-Ali je uporabnik tudi lastnik vozila"]
    newRow["C1.3-Obcina uporabnika vozila (opis)"] = row["C1.3-Obcina uporabnika vozila (opis)"]
    newRow["C2-Starost lastnika vozila"] = row["C2-Starost lastnika vozila"]
    newRow["C2-Ali je lastnik pravna ali fizicna oseba"] = row["C2-Ali je lastnik pravna ali fizicna oseba"]
    newRow["D.1-Znamka"] = row["D.1-Znamka"]
    newRow["D.3-Komerc. oznaka"] = row["D.3-Komerc. oznaka"]
    newRow["D.4.2-Drzava (koda)"] = row["D.4.2-Drzava (koda)"]
    newRow["G-Masa vozila"] = row["G-Masa vozila"]
    newRow["P.1.1-Delovna prostornina"] = row["P.1.1-Delovna prostornina"]
    newRow["P.1.2-Nazivna moc"] = row["P.1.2-Nazivna moc"]
    newRow["P.1.3-Vrsta goriva (opis)"] = row["P.1.3-Vrsta goriva (opis)"]
    newRow["P.1.3-Vrsta goriva (oznaka)"] = row["P.1.3-Vrsta goriva (oznaka)"]
    newRow["R-Barva vozila (opis)"] = row["R-Barva vozila (opis)"]
    newRow["S.1-Stevilo sedezev (vkljucno z vozniskim)"] = row["S.1-Stevilo sedezev (vkljucno z vozniskim)"]
    newRow["Okoljevarstvena oznaka"] = row["Okoljevarstvena oznaka"]
    newRow["Komerc. oznaka  do prvega /"] = row["Komerc. oznaka  do prvega /"]
    newRow["5A-Leto izdelave"] = row["5A-Leto izdelave"]
    
    
    data.append(newRow)

df = pd.DataFrame(data)

df.to_csv("vozila1.csv", encoding='utf-8')
```
